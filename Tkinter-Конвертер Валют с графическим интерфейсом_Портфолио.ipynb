```python
#Использование метод pack()

from tkinter import * #импорт библиотеки 

#создание главного окна в переменной root
root = Tk()

root.title("Currency Converter")# Изменение заголовка окна (не обязательно)
# Создание виджетов label для текстовых меток


label_1 = Label (root,text = "Input the Amount for Conversion:",
                font = "arial 12 bold", borderwidth = 10)

label_2 = Label (root,text = "Convert from Currency:",
                font = "arial 10 bold", borderwidth = 10)

label_3 = Label (root,text = "Convert to Currency:",
                font = "arial 10 bold", borderwidth = 10)

label_4 = Label (root,text = "Converted Amount:",
                font = "arial 10 bold", borderwidth = 10)

output = Label (root,text = "Print the OutPut here:",
                font = "arial 12 bold", borderwidth = 10)

#Создание виджетов entry для полей ввода
Enter_1 = Entry(root, borderwidth = 5)
Enter_2 = Entry(root, borderwidth = 5)
Enter_3 = Entry(root, borderwidth = 5)

#Сздание виджета Button для кнопки
my_button = Button(root, text = "CONVERT",
                  font = "arial 10", borderwidth = 5)

#Размешение виджетов в окне
label_1.pack()
label_2.pack()
label_3.pack()
label_4.pack()
output.pack()
Enter_1.pack()
Enter_2.pack()
Enter_3.pack()
my_button.pack()

#запускаем цикл , чтобы активировать графическый интерфейс
root.mainloop()
```

```python
# Код конвертация валют
import requests
    
#создание класса конвертера валют
class Currency_Converter:
    # словарь для хранения курсов валют
    rates = {}
    
    def __init__(self, url):
        # Запрос данных с URL-адрес или веб -страницы
        data = requests.get(url).json()
        
        # Извлечение курсов валют из полученных данных
        self.rates = data["rates"]
   
     # функция для выполнения комвертации
    def convert (self,from_currency, to_currency, amount):
        # т.к базовой валют является авро , то выполняем
        initial_amount = amount
        if from_currency != 'EUR':
            amount  = amount / self.rates[from_currency]
        output = amount* self.rates[to_currency]
        # Округляем преобразования до 2 знаков после запятой
        output = round(amount* self.rates[to_currency],2)
        # вывод в красивом формате для печати
        print('{}{} = {}{}'.format(initial_amount,
                    from_currency, output, to_currency))


if __name__=="__main__":
    
    # этот ключ у вас может не сработать,
    #  зарегриструйтесь и получите свой собственый ключ доступа
    ACCESS_KEY="6d16f56a7cfcb160d2759ed48d699e3f"
    url = str.__add__('http://data.fixer.io/api/latest?access_key=', 
                      ACCESS_KEY)
    # вызов экземпляра класса
    converter = Currency_Converter(url)
    # Получение даннхы от пользователя
    from_country  = input ("converte from Currency:")
    to_country  = input ("converte to Currency:")
    amount  = int(input ("Amount to beconverted:"))
    
    converter.convert(from_country, to_country, amount)
    
            
```

```python
#Итоговое прилжение

#импорт библиотеки 
import tkinter as tk 

window = tk.Tk()

window.title("Currency Converter")

Input_label = tk.Label(window, 
                      text = "Input the Amount for  Conversion:",
                      font = "arial 12 bold ", borderwidth = 10)
label_fc = tk.Label(window, 
                      text = "Convert from Currency:",
                      font = "arial 12 bold ", borderwidth = 10)
label_tc = tk.Label(window, 
                      text = "Convert to Currency:",
                      font = "arial 12 bold ", borderwidth = 10)
outputlabel = tk.Label(window, 
                      text = "Convert to Amount:",
                      font = "arial 12 bold ", borderwidth = 10)

input_amount = tk.Entry(window,font = "arial 12 bold", 
                       borderwidth = 5)
from_currency = tk.Entry(window,font = "arial 12 bold", 
                       borderwidth = 5)
to_currency = tk.Entry(window,font = "arial 12 bold", 
                       borderwidth = 5)
output = tk.Entry(window,font = "arial 12 bold", 
                       borderwidth = 5)

Input_label.grid(row = 0,column = 0)
label_fc.grid(row =1,column = 0)
label_tc.grid(row = 2,column = 0)

input_amount.grid(row = 0,column = 1)
from_currency.grid(row =1,column = 1)
to_currency.grid(row = 2,column = 1)

outputlabel.grid(row = 4,column = 0)
output.grid(row = 4,column = 1)

ACCESS_KEY="6d16f56a7cfcb160d2759ed48d699e3f"
url = str.__add__('http://data.fixer.io/api/latest?access_key=', 
                      ACCESS_KEY)

# Код конвертация валют
import requests

# Запрос данных с URL-адрес или веб -страницы
data = requests.get(url).json()
        
# Извлечение курсов валют из полученных данных
rates = data["rates"]
   
# функция для выполнения комвертации
def currency_converter (from_currency, to_currency, input_amount):
    from_currency = from_currency.get()
    to_currency = to_currency.get()
    
    input_amount = float(input_amount.get())
    if from_currency != 'EUR':
        input_amount  = input_amount / rates[from_currency]
    # Округляем преобразования до 2 знаков после запятой
    output.insert(0, round(input_amount* rates[to_currency],2))
    return


convert_button = tk.Button(window, text = "CONVERT",
                font="arial 16 bold", borderwidth = 10,
                padx = 50,bg = "powder blue",
                command = lambda: currency_converter(from_currency, 
                to_currency, input_amount))

convert_button.grid(row = 3, columnspan = 2)
window.mainloop()
    
```
